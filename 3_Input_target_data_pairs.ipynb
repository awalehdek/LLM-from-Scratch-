# Input & target pairs are training example. Used to teach the model to predict target
input: data given 
target: expected output


1-Type

A- Next token prediction

Input: All tokens except the last one

Target: The next token in sequence

text = "The cat sat on the mat"

tokens = ["The", "cat", "sat", "on", "the", "mat"]

 Training pairs:
input  = ["The"]                    → target = ["cat"]

input  = ["The", "cat"]             → target = ["sat"]  

input  = ["The", "cat", "sat"]      → target = ["on"]

input  = ["The", "cat", "sat", "on"] → target = ["the"]

B- Sequence To Sequence:

        -Input: Source language text
        -Target: Target language translation

input  = "I love machine learning"

target = "J'aime l'apprentissage automatique"

C. Text Completion
    
    - Input: Partial text
    -Target: Completion text

input  = "The weather today is"

target = "sunny and warm"

D. Classification

    - Input: Text to classify  
    - Target: Category/label

input  = "This movie is fantastic and well-acted"

target = "positive"
