# Input & target pairs are training example. Used to teach the model to predict target
input: data given 
target: expected output


### 1-Type


A- Next token prediction

Input: All tokens except the last one

Target: The next token in sequence

text = "The cat sat on the mat"

tokens = ["The", "cat", "sat", "on", "the", "mat"]

 Training pairs:
input  = ["The"]                    → target = ["cat"]

input  = ["The", "cat"]             → target = ["sat"]  

input  = ["The", "cat", "sat"]      → target = ["on"]

input  = ["The", "cat", "sat", "on"] → target = ["the"]

B- Sequence To Sequence:

        -Input: Source language text
        -Target: Target language translation

input  = "I love machine learning"

target = "J'aime l'apprentissage automatique"

C. Text Completion
    
    - Input: Partial text
    -Target: Completion text

input  = "The weather today is"

target = "sunny and warm"

D. Classification

    - Input: Text to classify  
    - Target: Category/label

input  = "This movie is fantastic and well-acted"

target = "positive"

```python
import torch

def create_input_target_pairs(text_tokens, sequence_length=4):
    """
    Create (input, target) pairs for next token prediction
    """
    inputs = []
    targets = []
    
    for i in range(len(text_tokens) - sequence_length):
        # Input: sequence of tokens
        input_seq = text_tokens[i:i + sequence_length]
        # Target: next token after the sequence
        target_token = text_tokens[i + sequence_length]
        
        inputs.append(input_seq)
        targets.append(target_token)
    
    return inputs, targets

# Example usage
text = "The quick brown fox jumps over the lazy dog"
tokens = text.split()  # ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']

inputs, targets = create_input_target_pairs(tokens, sequence_length=3)

print("Input -> Target pairs:")
for i in range(len(inputs)):
    print(f"{inputs[i]} -> {targets[i]}")
```
